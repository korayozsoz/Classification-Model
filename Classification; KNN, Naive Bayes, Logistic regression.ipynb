```python
import pandas as pd
import numpy as np
```

```python
telco=pd.read_csv("Telco.csv",na_values=" ")
```

```python
telco.info()
```

```python
telco.drop( "customerID", axis=1,inplace=True)
```

```python
# I will assign "numeric values" for our variables that are categorical.
dummies=pd.get_dummies(telco,drop_first=True)
dummies_without_na=dummies.dropna()
```

```python
dummies_without_na.head()
```

```python
X=dummies_without_na.drop(["Churn_Yes"],axis=1)
y=dummies_without_na.Churn_Yes
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# So, that there is no data snooping in our model
# I split the dataset train(0.8) and test (0,2).
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2, random_state=2, stratify=y)
```

**Logistic Regression**

```python
model=LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
```

```python
ypred=model.predict(X_test)
ypred
```

```python
y.shape
```

```python
X.shape
```

```python
X
```

```python
score=model.score(X_test, y_test)
print("Score of the model is: ",score)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score,precision_score, classification_report, f1_score
```

```python
ypred=model.predict(X_test)
ypred
```

```python
print("Accuracy Score: ",accuracy_score(y_test,ypred))
print("Confusion Matrix: \n",confusion_matrix(y_test,ypred))
print("Recall Score: ",recall_score(y_test,ypred))
print("Precision Score: ",precision_score(y_test,ypred))
print("F1 Score: ",f1_score(y_test,ypred))
print("-----------------------------------------------------------------------------")
print(classification_report(y_test,ypred))
```

**K-NN**

```python
from sklearn.neighbors import KNeighborsClassifier
```

-->I will investigate Whick k value best for the model (for: 1<k<#of_features)



```python
liste=[]
for k in range(1,X.shape[1]):
  model=KNeighborsClassifier(n_neighbors=k)
  model.fit(X_train,y_train)
  liste.append([model.score(X_test,y_test),k])
df=pd.DataFrame(liste, columns=["Score","Neighbours_number"])
print(df)
```

```python
df.Score.max()
```

--> Appearently, if neighbour number is 18, score of our model is more higher!!



---> Now I will change the metrics while n_neighbours remaining same (18).


```python
  model=KNeighborsClassifier(n_neighbors=18, metric="euclidean")
  model.fit(X_train,y_train)
```

```python
a=[]
metric_liste=["manhattan","euclidean","cosine"]
for i in metric_liste:
  model=KNeighborsClassifier(n_neighbors=6, metric=i)
  model.fit(X_train,y_train)
  model.score(X_test,y_test)
  a.append([i,model.score(X_test,y_test)])
dff=pd.DataFrame(a, columns=["Metric","Score"])
dff
```

```python
dff.Score.max()
```

--> Appearently, **cosine** distance best for our KNN classification model.
